Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Apply a function to check the value of `class` column
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

```python
# 1. Initialiser le MarkerCluster
marker_cluster = MarkerCluster()

# 2. Ajouter le marker_cluster à la carte principale (site_map)
site_map.add_child(marker_cluster)

# 3. Parcourir chaque ligne du DataFrame spacex_df
for index, record in spacex_df.iterrows():
    # Extraction des coordonnées du lancement actuel
    coordinate = [record['Lat'], record['Long']]
    
    # Création du Marqueur individuel
    marker = folium.Marker(
        location=coordinate,
        # Personnalisation de la couleur de l'icône selon la réussite (vert) ou l'échec (rouge)
        icon=folium.Icon(color='white', icon_color=record['marker_color'], icon='info-sign'),
        # Optionnel: afficher le nom du site et le résultat au clic
        popup=f"Site: {record['Launch Site']}<br>Status: {'Success' if record['class'] == 1 else 'Failed'}"
    )
    
    # Ajouter le marqueur au cluster plutôt qu'à la carte directement
    marker_cluster.add_child(marker)

# 4. Afficher la carte mise à jour
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
def calculate_distance(lat1, lon1, lat2, lon2):
    # Rayon de la Terre en kilomètres
    R = 6373.0

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Coordonnées trouvées sur la carte pour la côte la plus proche
coastline_lat = 28.56367
coastline_long = -80.57163

# Calcul de la distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_long)
print(f"Distance à la côte : {distance_coastline:.2f} KM")
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
# 1. Définir le point de la côte
coast_coordinate = [coastline_lat, coastline_long]
launch_coordinate = [launch_site_lat, launch_site_lon]

# 2. Créer et ajouter le marqueur de distance sur la côte
distance_marker = folium.Marker(
    location=coast_coordinate,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12px; color:#d35400; width: 150px;"><b>{distance_coastline:.2f} KM</b></div>',
    )
)
site_map.add_child(distance_marker)

# 3. Tracer une ligne (PolyLine) entre le site de lancement et la côte
# Cela permet de visualiser la trajectoire la plus courte vers l'océan
lines = folium.PolyLine(
    locations=[launch_coordinate, coast_coordinate], 
    weight=2,        # Épaisseur de la ligne
    color='blue'     # Couleur de la ligne
)
site_map.add_child(lines)

# 4. Afficher la carte mise à jour
site_map
```
